üåü Exercise 7 :
Instructions
Create a line plot and scatter plot representing your favorite mathematical functions. 
Change the line color to tab:blue for the line plot and use a marker of your choice for the scatter plot.
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

```python
x1 = np.linspace(-10, 10, 1000)
x2 = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10, 3))
ax1.plot(x1, y1, color='tab:blue')
ax2.scatter(x2, y2, marker='*')

ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('sin(x)')

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('cos(x)')

plt.show()
```
